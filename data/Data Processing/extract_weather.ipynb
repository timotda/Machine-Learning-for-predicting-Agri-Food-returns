URL to download raw data: https://www.ncei.noaa.gov/access/search/data-search/daily-summaries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

import extraction
# Definitions of all functions that will be used in this notebook are to be found in 'extraction.py'.
# This notebook is exclusively used for visualization and finally saving processed data.

base_dir = os.getcwd()
print("Current WD: ", base_dir)

#raw_data_dir = os.path.join(base_dir, "raw data weather")
raw_data_dir = "data/Data Files/raw data weather"
```

```python
columns_to_keep = ['DATE',
                   'TMIN', 'TMAX', # temperature min, max
                   'PRCP', # precipitation
                   'AWND', # avg wind speed
                   'SNOW', 'SNWD' # snow, snow depth
                   ]

global_start_date = pd.to_datetime('01/01/1930', format='%d/%m/%Y')
global_end_date = pd.to_datetime('31/12/2025', format='%d/%m/%Y')
full_date_index = pd.date_range(start=global_start_date, end=global_end_date, freq='D')
```

```python
weather_files = [f for f in os.listdir(raw_data_dir) if f.endswith('_weather_data.csv')]

dfs = []
for file in weather_files:
    state = file.split('_')[0]
    print(state)
    path = os.path.join(raw_data_dir, file)
    df_state = extraction.load_and_process_file(path, state, columns_to_keep, full_date_index)
    dfs.append(df_state)
```

```python
df_all_states = pd.concat(dfs, axis=1, join='outer')
df_all_states.index.name = 'DATE'

display(df_all_states)
```

```python
extraction.nan_summary(df_all_states)
```

```python
df_cut = extraction.cut_data(df_all_states, '2000-01-01', '2024-12-31')
display(df_cut)
```

```python
extraction.nan_summary(df_cut)
```

```python
df_cut = df_cut.interpolate()

# save to current wd
output_path = os.path.join(os.getcwd(), "weather_data_00_24.csv")
df_cut.to_csv(output_path)
```
